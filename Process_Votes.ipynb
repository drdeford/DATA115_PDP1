```python
import pandas as pd
```

```python
df = pd.read_csv("./Raw_Data/S114_votes.csv")
```

```python
dfM = pd.read_csv("./Raw_Data/S114_members.csv")
```

```python
df.head()
```

```python
dfM.head()
```

```python
votes = []
hvotes=[]
ids = df['icpsr'].unique()
for id in ids: 
    votes.append([id,str(dfM[dfM['icpsr']==id]['bioname']).split()[1][:-1],int(dfM[dfM['icpsr']==id]['party_code'])])
    for i in range(500):
        votes[-1].append(list(df[df['icpsr']==id]['cast_code'])[i])
```

```python
votes[1]
```

```python
dfvote = pd.DataFrame(votes)
```

```python
dfvote.head()
```

```python
dfvote=dfvote.rename(columns={0: "id", 1: "Name", 2: "Party"})
```

```python
dfvote.to_csv("./Processed_Data/Processed_Senate_Data.csv")
```

```python
hvotes=[]
ids = df['icpsr'].unique()
for id in ids: 
    hvotes.append(list(df[df['icpsr']==id]['cast_code']))
```

```python
from scipy.spatial.distance import hamming
import numpy as np
```

```python
hamming(hvotes[1],hvotes[1])
```

```python
D = np.zeros([100,100])
for i in range(100):
    for j in range(100):
        D[i,j] = hamming(hvotes[i],hvotes[j])
```

```python
D
```

```python
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
```

```python
Xhat = MDS(n_components = 2, dissimilarity = 'precomputed').fit_transform(D)
```

```python
cdict={200:'red',100:'blue',328:'green'}
plt.scatter(Xhat[:,0],Xhat[:,1],c=[cdict[x] for x in dfvote['Party']])
```

```python
Xhat = MDS(n_components = 2).fit_transform(dfvote[[x for x in range(4,400)]])
```

```python
plt.scatter(Xhat[:,0],Xhat[:,1],c=dfvote['Party'])
```
